# create scatterplot of"sepalalegthem 'vs' sepalwidthem

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Iris.csv")

# Clean column names (recommended)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Basic scatter plot
sns.scatterplot(data=df, x="sepallengthcm", y="sepalwidthcm",hue='species')

plt.title("Sepal Length vs Sepal Width")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.show()
```

#  14 create scatterplot of"sepalalegthem 'vs' sepalwidthem using seaborn's facegrid

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("Iris.csv")

# Clean up column names: strip spaces, lowercase, replace spaces with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')


# FacetGrid scatter plot
g = sns.FacetGrid(df, col="species", height=4)
g.map(sns.scatterplot, "sepallengthcm", "sepalwidthcm")  # use lowercase
g.fig.suptitle("Sepal Length vs Sepal Width by Species", y=1.05)
plt.show()
```

# Boxplot of petalwidth for diffferent species

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Iris.csv")

# Clean column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Create the boxplot
sns.boxplot(data=df, x='species', y='petalwidthcm')

plt.title("Petal Width by Species")
plt.xlabel("Species")
plt.ylabel("Petal Width (cm)")
plt.show()
```
